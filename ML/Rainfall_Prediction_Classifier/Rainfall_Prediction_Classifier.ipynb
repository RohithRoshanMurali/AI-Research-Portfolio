## Project

```python
# Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
```

```python
# Loading and Inspecting Data

url="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/_0eYOqji3unP1tDNKWZMjg/weatherAUS-2.csv"
df = pd.read_csv(url)
df.head()
```

```python

df.count()
```

```python
# Drop all the missing values

df = df.dropna()
df.info()
```

```python
# Since we have 56k entries, we dont have the need to impute any missing values

df.columns
```

```python
# Update the column names to avoid confusion

df = df.rename(columns={'RainToday' : 'RainYesterday', 'RainTomorrow' : 'RainToday'})
```

```python
# Location selection
df = df[df.Location.isin(['Melbourne','MelbourneAirport','Watsonia',])]
df. info()
```

```python
# We still have 7k entries which is good to build a reasonable model.
```

```python
# Extracting seasonlaity feature

def date_to_season(date):
  month = date.month
  if(month == 12) or (month== 1) or (month == 2):
    return 'Summer'
  elif(month == 3) or (month == 4) or (month == 5):
        return 'Autumn'
  elif (month == 6) or (month == 7) or (month == 8):
        return 'Winter'
  elif (month == 9) or (month == 10) or (month == 11):
        return 'Spring'
```

```python
# Mapping dates to seasons and dropping Date column

df['Date'] = pd.to_datetime(df['Date'], errors = 'coerce')
df['Season'] = df['Date'].apply(date_to_season)
df = df.drop(columns=['Date'])
df.head()
```

```python
# Define the feature and target variables

X = df.drop(columns = 'RainToday', axis =1)
y = df['RainToday']
```

```python
# Check the balance of the class

y.value_counts()
```

```python
# Since the data isnt balanced, we will be using stratified train test split

X_train, X_test, y_train, y_test = train_test_split(
    X,y,
    test_size = 0.2,
    stratify = y,
    random_state = 42
)
```

```python
y_train.value_counts(normalize=True)
y_test.value_counts(normalize=True)
```

```python
# Separating numerical and categorical features

numeric_features = X_train.select_dtypes(['float64', 'int64']).columns.tolist()
categorical_features = X_train.select_dtypes('object', 'category').columns.tolist()

# Defining seprate transformers and combining them into single transformer

numeric_transformer = Pipeline(steps=[('scaler', StandardScaler())])
categorical_transformer = Pipeline(steps=[('onehot', OneHotEncoder(handle_unknown='ignore'))])

# Combining into a single transformer

from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(
    transformers = [
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)
```

```python
# Pipeline combining preprocessing and Random Classifier

pipeline = Pipeline(
    steps = [
        ('preprocessor', preprocessor),
        ('classifier', RandomForestClassifier(random_state=42))
    ]
)
```

```python
# Defining parameter grid to use cross-validation grid search in validation

param_grid = {
    'classifier__n_estimators' : [ 50,100 ],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5]
}
```

```python
# CV method

cv = StratifiedKFold(n_splits=5, shuffle=True)
```

```python
# Fit gridsearchcv to pipeline

grid_search = GridSearchCV(pipeline, param_grid, cv=cv, scoring = 'accuracy', verbose =2)
grid_search.fit(X_train, y_train)
```

```python
# Print the best parameter and CV score

print('\n Best parameters found : ', grid_search.best_params_)
print('\nBest CV score : {:.2f}'.format(grid_search.best_score_) )
```

```python
# Model's estimated score

test_score = grid_search.score(X_test, y_test)
print('Test score : {:.2f}'.format(test_score))
```

```python
# Model predictions for unseen data

y_pred = grid_search.predict(X_test)
```

```python
# Classification report

print('\n Classification report:')
print(classification_report(y_test, y_pred))
```

```python
# Plotting confusion matrix

conf_matrix = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix)
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Extracting the important features

feature_importances = grid_search.best_estimator_['classifier'].feature_importances_
```

```python
# Combining numerical and categorical feature names

feature_names = numeric_features + list(
    grid_search.best_estimator_['preprocessor']
    .named_transformers_['cat']
    .named_steps['onehot']
    .get_feature_names_out(categorical_features)
)

feature_importances = grid_search.best_estimator_['classifier'].feature_importances_

importance_df = pd.DataFrame({
    'Feature' : feature_names,
    'Importance' : feature_importances
}).sort_values(by='Importance', ascending = False)

N = 10  # Number of features to be displayed
top_features = importance_df.head(N)


```

```python
# Plotting

plt.figure(figsize=(10, 6))
plt.barh(top_features['Feature'], top_features['Importance'], color='skyblue')
plt.gca().invert_yaxis()  # Invert y-axis to show the most important feature on top
plt.title(f'Top {N} Most Important Features in predicting whether it will rain today')
plt.xlabel('Importance Score')
plt.show()
```
